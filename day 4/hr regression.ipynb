# Regression

```python
miss = hr.isnull().sum()
miss[miss>0]
```

```python
data = hr.copy()
del data["Attrition"]
```

```python
data.isnull().sum()
```

```python
missing = data[data["MonthlyIncome"].isnull()]
filled = data[data["MonthlyIncome"].notnull()]
```

```python
missing.shape, filled.shape
```

```python
filled.corr()
```

```python
abs(filled.corr()["MonthlyIncome"]).sort_values(ascending=False)
```

```python
from scipy import stats
est = stats.linregress(x=filled["TotalWorkingYears"], y=filled["MonthlyIncome"])
```

```python
est
```

```python
est.slope
```

```python
est.intercept
```

```python
def predict(x):
    return est.slope * x + est.intercept
```

```python
predict(2)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2_score(filled["MonthlyIncome"], filled["TotalWorkingYears"].apply(predict))
```

```python
mean_squared_error(filled["MonthlyIncome"], filled["TotalWorkingYears"].apply(predict)) ** .5
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(filled["TotalWorkingYears"], filled["MonthlyIncome"])
plt.plot(filled.TotalWorkingYears, predict(filled.TotalWorkingYears), color="orange")
```

```python
filled.info()
```

```python
filled.select_dtypes(exclude="number").head()
```

```python
filled.select_dtypes(exclude="number").nunique()
```

```python
X = pd.get_dummies(filled, drop_first=True)
X.info()
```

```python
X.head()
```

```python
from sklearn.linear_model import LinearRegression
L = LinearRegression()
```

```python
L.fit(X.drop("MonthlyIncome", axis=1), X.MonthlyIncome);
```

```python
pred = L.predict(X.drop("MonthlyIncome", axis=1))
pred
```

```python
r2_score(X["MonthlyIncome"], pred)
```

```python
mean_squared_error(X["MonthlyIncome"], pred) ** 0.5
```

# Ridge

```python
from sklearn.linear_model import Ridge
R = Ridge()
```

```python
R.fit(X.drop("MonthlyIncome",axis=1), X["MonthlyIncome"]);
```

```python
r2_score(X["MonthlyIncome"], R.predict(X.drop("MonthlyIncome", axis=1)))
```

```python
mean_squared_error(X["MonthlyIncome"], R.predict(X.drop("MonthlyIncome", axis=1))) ** 0.5
```

# Lasso

```python
from sklearn.linear_model import Lasso
Las = Lasso(alpha=100)
```

```python
Las.fit(X.drop("MonthlyIncome",axis=1), X["MonthlyIncome"]);
```

```python
r2_score(X["MonthlyIncome"], Las.predict(X.drop("MonthlyIncome", axis=1)))
```

```python
mean_squared_error(X["MonthlyIncome"], Las.predict(X.drop("MonthlyIncome", axis=1))) ** 0.5
```

# Elastic-Net

```python
from sklearn.linear_model import ElasticNet
E = ElasticNet(l1_ratio=.8)
```

```python
E.fit(X.drop("MonthlyIncome", axis=1), X["MonthlyIncome"]);
```

```python
r2_score(X["MonthlyIncome"], E.predict(X.drop("MonthlyIncome", axis=1)))
```
