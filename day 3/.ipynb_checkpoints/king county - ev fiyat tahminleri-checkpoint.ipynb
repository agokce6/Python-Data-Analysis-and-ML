```python
x = df[["Mileage","Cylinder","Liter","Cruise"]] # cor>20 and cor<90
```

```python
x.head()
```

```python
y = df[["Price"]]
y
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=13)

print("Values in x train", x_train)
print("Values in y train", y_train)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
lr = LinearRegression()
regmodel = lr.fit(x_train, y_train)
```

```python
y_pretest = regmodel.predict(x_test)
y_pretest
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
(mean_squared_error(y_test, y_pretest)) ** 0.5
```

```python
r2_score(y_test, y_pretest)
```

```python
x, y = df[["Mileage", "Cylinder", "Liter", "Cruise", "Make", "Trim"]], df[["Price"]]
```

```python
x.head()
```

```python
x = pd.get_dummies(x, drop_first=True)
x.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 13)
lr = LinearRegression()
regmodel = lr.fit(x_train, y_train)
y_pretest = regmodel.predict(x_test)
```

```python
(mean_squared_error(y_test, y_pretest)) ** 0.5
r2_score(y_test, y_pretest)
```

```python
# LR
from sklearn.linear_model import Ridge, Lasso
```

```python
R = Ridge()
R.fit(x_train, y_train)
y_pred_ridge = R.predict(x_test)
```

```python
y_pred_ridge
```

```python
(mean_squared_error(y_test, y_pred_ridge)) ** 0.5
```

```python
r2_score(y_test, y_pred_ridge)
```

```python
# Lasso
L = Lasso()
L.fit(x_train, y_train)
y_pred_lasso = L.predict(x_test)
```

```python
(mean_squared_error(y_test, y_pred_lasso)) ** 0.5
```

```python
r2_score(y_test, y_pred_lasso)
```

```python
from sklearn.linear_model import ElasticNet
E = ElasticNet(l1_ratio= 0.8)
```

```python
E.fit(x_train, y_train)
y_pred_elastic = E.predict(x_test)
```

```python
(mean_squared_error(y_test, y_pred_elastic)) ** 0.5
```

```python
r2_score(y_test, y_pred_elastic)
```
