```python
from gtts import gTTS
myText = """ merhaba ahmet bey, yapay zeka dersimize hoşgeldiniz. ben sizin özel asistanınızım"""

output = gTTS(text=myText, lang='tr', slow=False)
output.save("outputtr.mp3")
```

```python
import os, io
fh = io.open("text.txt","r", encoding="utf-8")
myText2 = fh.read().replace("\n", " ")

language = "tr"

output = gTTS(myText2, lang=language, slow=False)
output.save("output.mp3")
fh.close()

os.system("start output.mp3")
```
